```python
import pandas as pd #manipular nos dados
import matplotlib.pyplot as plt #criar tabela/gráficos
import seaborn as sns #exibição dos gráficos
```

```python
dados = pd.read_excel("/content/drive/MyDrive/Faculdade/3º semestre/Sistemas de Apoio à decisão/dados-12-14.xls")
```

```python
dados2 = dados.set_index("dateRep")
dados2 = dados2.sort_index()
dados2
```

```python
dados3 = dados2.pivot(columns = "countriesAndTerritories")
dados3
```

```python
dados3.columns
```

```python
dados3.head()
```

```python
sns.set(rc={'figure.figsize':(20,10)})
dados3['cases'].plot()
```

```python
dados3['cases'].plot(logy=True)
```

```python
dados3['cases'].cumsum().plot()
```

```python
dados2["cum_cases"]= dados2.groupby('countriesAndTerritories')['cases'].cumsum()
```

```python
dados4 = dados2.pivot(columns='countriesAndTerritories')
dados4.columns
```

```python
dados4['cum_cases'].plot()
```

```python
dados4['cum_cases'].plot(logy=True)
```

```python
(dados4['cum_cases']/dados4['popData2018']).plot()
```

```python
dados2['cases_per_pop'] = dados2['cum_cases'] / dados2['popData2018']
```

```python
dados5 = dados2.pivot(columns = 'countriesAndTerritories')
dados5['cases_per_pop'].plot()
```
